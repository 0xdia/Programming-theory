```python
import pygraphviz as pgv
from IPython.display import SVG
from main import articulation_points
```

```python
g = pgv.AGraph("./graph0.gv").to_undirected()
g.node_attr["color"] = "green"
g.node_attr["label"] = " "
aps = articulation_points(g)
for ap in aps:
    ap.attr["fillcolor"] = "red"
g.layout()
g.draw("result.svg")
SVG("./result.svg")
```
